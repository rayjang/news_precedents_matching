```python
import os
import glob
  
# use glob to get all the csv files 
# in the folder
path = os.getcwd() + "/naver-news-data/"
csv_files = glob.glob(os.path.join(path, "*.csv"))
  
# csv_files = ['/mnt/workspace/MyCanvas/naver-news-data/NaverNews_공직선거법 제230조 제1항 제4호.csv']

# loop over the list of csv files
for f in csv_files:
    statute = f.split("/")[-1].split(".")[0].split("_")[1]
    idx_list = [i for i in range(len(id_list)) if statute in statutes_list[i]]
    # read the csv file
    df = pd.read_csv(f)

    column_names = ["title", "link", "content", "law_id", "casetype", "casename", "statutes", "facts"]
    new_df = pd.DataFrame(columns = column_names)
   
    for i in range(df.shape[0]):
        _title = df.loc[i, "title"].replace('[','').replace(']','')
        _content = df.loc[i, "content"].replace('[','').replace(']','')
        if len(_title) < 5 or len(_content) < 20:
            continue
        for j in idx_list:
            df.loc[i,"title"] = df.loc[i,"title"].replace('[','').replace(']','').replace('\n','')
            df.loc[i,"content"] = df.loc[i,"content"].replace('[','').replace(']','').replace('\n','')
            new_df.loc[len(new_df)] = [df.loc[i,"title"], df.loc[i,"link"], df.loc[i,"content"], id_list[j], casetype_list[j], casename_list[j], statutes_list[j], facts_list[j]]
    if new_df.shape[0] > 1:
        new_df.to_csv(f'./news-law-data/{statute}.csv', index=False, encoding='utf-8')
```

```python
import os
import glob
  
# use glob to get all the csv files 
# in the folder
path = os.getcwd() + "/naver-news-data/"
csv_files = glob.glob(os.path.join(path, "*.csv"))
  
texts = []
labels = []

# loop over the list of csv files
for f in csv_files:
    statute = f.split("/")[-1].split(".")[0].split("_")[1]
    idx_list = [i for i in range(len(id_list)) if statute in statutes_list[i]]
    # read the csv file
    df = pd.read_csv(f)
    
    for i in range(df.shape[0]):
        _title = df.loc[i, "title"].replace('[','').replace(']','')
        _content = df.loc[i, "content"].replace('[','').replace(']','')
        if len(_title) < 5 or len(_content) < 20:
            continue
        
        df.loc[i,"title"] = df.loc[i,"title"].replace('[','').replace(']','').replace('\n','')
        df.loc[i,"content"] = df.loc[i,"content"].replace('[','').replace(']','').replace('\n','')
        
        texts.append(df.loc[i,"title"] + '\n' + df.loc[i,"content"])
        labels.append(statute)
    
    for j in idx_list:
        texts.append(facts_list[j])
        labels.append(statute)
```

```python
import random

c = list(zip(texts, labels))

random.shuffle(c)

texts, labels = zip(*c)
```

```python
import pickle
with open('texts_list.pickle', 'wb') as f:
    pickle.dump(texts, f, pickle.HIGHEST_PROTOCOL)
with open('labels_list.pickle', 'wb') as f:
    pickle.dump(labels, f, pickle.HIGHEST_PROTOCOL)
```
